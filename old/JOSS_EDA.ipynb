```python
import glob
import json
import pandas as pd
```

```python
input_path = 'input/'
input_extension = '.trf'
files = glob.glob(input_path + '**/*' + input_extension, recursive=True)
with open('input/JOSS/support/variables_info.json') as f:
    variables_info = json.load(f)
```

```python
variables_info.keys()
```

```python
df = pd.read_table(files[0], delimiter=r"\s+")
```

```python
# Read all files and concatenate them
df = pd.concat([pd.read_table(f, delimiter=r"\s+", decimal=",") for f in files])
df['timestamp'] = pd.to_datetime(df['YYYY-MM-DD'] + ' ' + df['hh:mm:ss'])
df = df.drop(['YYYY-MM-DD', 'hh:mm:ss'], axis=1)
df = df.set_index('timestamp')
df = df[~df.index.duplicated(keep='first')]
df = df.sort_index()
```
