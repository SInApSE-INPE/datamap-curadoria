```python
import xarray as xr
```

```python
ds = xr.open_mfdataset('output/netCDF/*.nc', combine='nested', concat_dim='time', parallel=True)
```

```python
ds
```

```python
ds['number_detected_drops'].plot(figsize=(18, 2))
ds['diam_max'].plot(figsize=(18, 2))
ds['diam_min'].plot(figsize=(18, 2))
ds['rain_rate'].plot(figsize=(18, 2))
ds['radar_reflectivity'].plot(figsize=(18, 2))
ds['liq_water'].plot(figsize=(18, 2))
ds['energy_flux'].plot(figsize=(18, 2))
ds['slope_parameter'].plot(figsize=(18, 2))
ds['distribution_intercept'].plot(figsize=(18, 2))
```

```python
for avg_c in range(ds['fall_velocity'][0,:].size):
    ds['fall_velocity'][:,avg_c].plot(figsize=(18, 2))

for avg_c in range(ds['delta_diam'][0,:].size):
    ds['delta_diam'][:,avg_c].plot(figsize=(18, 2))

for avg_c in range(ds['num_drop'][0,:].size):
    ds['num_drop'][:,avg_c].plot(figsize=(18, 2))
```
